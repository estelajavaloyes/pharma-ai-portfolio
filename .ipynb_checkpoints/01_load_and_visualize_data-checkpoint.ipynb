```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

print("os, numpy, matplotlib y PIL cargados correctamente")
```

```python
data_dir = "data"
```

```python
classes = os.listdir(data_dir)
print("Clases encontradas:", classes)
```

```python
plt.figure(figsize=(8,8))

for i, cls in enumerate(classes):
    cls_path = os.path.join(data_dir, cls)
    img_name = os.listdir(cls_path)[0] 
    img = Image.open(os.path.join(cls_path, img_name))
    plt.subplot(2,2,i+1)
    plt.imshow(img)
    plt.title(cls) 
    plt.axis('off')

plt.show()
```

```python
IMG_SIZE = 128

X = []
y = []

for idx, cls in enumerate(classes):
    cls_path = os.path.join(data_dir, cls)
    
    for img_name in os.listdir(cls_path):
        img_path = os.path.join(cls_path, img_name)

        img = Image.open(img_path)

        img = img.convert('L')

        img = img.resize((IMG_SIZE, IMG_SIZE))

        img_array = np.array(img)

        X.append(img_array)
        y.append(idx)

X = np.array(X)
y = np.array(y)

print("Shape de X:", X.shape)
print("Shape de y:", y.shape)
print("Mapa de clases:", {i: c for i, c in enumerate(classes)})
```

```python
from sklearn.model_selection import train_test_split

X_flat = X.reshape(X.shape[0], -1)

X_train, X_test, y_train, y_test = train_test_split(
    X_flat, y, test_size=0.2, random_state=42, stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=classes))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=classes,
            yticklabels=classes)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

## Clinical Interpretation

The confusion matrix shows that the model performs well across most classes, particularly for Normal and Very Mild Dementia cases.

However, the dataset is highly imbalanced, with very few Moderate Demented cases. This may limit the model’s ability to generalize to severe stages of the disease.

From a clinical perspective, the model shows strong performance in early-stage detection (Very Mild and Mild Dementia), which is particularly relevant for early diagnosis and potential therapeutic intervention.

Limitations:
- Class imbalance
- Dataset size
- Logistic Regression is a simple baseline model

Future work:
- Use deep learning (CNN) for feature extraction
- Apply data augmentation
- Evaluate on external datasets

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report

accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", accuracy_rf)

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, target_names=classes))
```

## Model Comparison

Logistic Regression outperformed Random Forest on this dataset (96.5% vs 92.0% accuracy).

Random Forest showed poor performance for the Moderate Demented class (recall = 0.23), likely due to the severe class imbalance (only 13 samples in the test set).

This highlights an important limitation of the dataset: models may perform well overall but fail to detect clinically severe cases.

From a clinical perspective, this is critical, as missing moderate-stage patients could impact treatment decisions.

Future improvements:
- Address class imbalance (class weighting or resampling)
- Collect more samples for severe stages
- Evaluate deep learning approaches

```python
misclassified_idx = np.where(y_test != y_pred)[0]

print("Número de errores:", len(misclassified_idx))
```

```python
plt.figure(figsize=(10,8))

for i, idx in enumerate(misclassified_idx[:6]):
    img = X_test[idx].reshape(128,128)
    true_label = classes[y_test[idx]]
    pred_label = classes[y_pred[idx]]
    
    plt.subplot(2,3,i+1)
    plt.imshow(img, cmap='gray')
    plt.title(f"True: {true_label}\nPred: {pred_label}")
    plt.axis('off')

plt.show()
```

## Error Analysis

Several misclassified cases were observed, particularly between adjacent disease stages (e.g., Very Mild vs Mild Dementia).

This is clinically plausible, as early stages of Alzheimer’s disease often show subtle structural differences in MRI scans.

Such errors highlight the difficulty of stage differentiation and the need for more advanced models or multimodal data.
